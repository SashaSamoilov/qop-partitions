```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

matplotlib.rcParams.update({'font.size': 11})
```

```python
fig, ax = plt.subplots(figsize=(6, 4), dpi=300)
ax.set_xlim([1.75, 8.25])
for axis in [ax.xaxis]:
    axis.set_major_locator(ticker.MaxNLocator(integer=True))

x = [2, 4, 6, 8]
t1 = 2.094
t2 = 38.255
t3 = 530.554
t4 = 6091.733
z1 = [t1/1.672, t1/0.958, t1/0.958, t1/0.973]    
z2 = [t2/30.770, t2/15.684, t2/14.721, t2/14.636] 
z3 = [t3/308.938, t3/156.577, t3/147.153, t3/146.164]
z4 = [t4/3920.940, t4/2323.431, t4/2300.573, t4/2309.608]

ax.set_xlabel('Число используемых процессоров')
ax.set_ylabel('Ускорение')
plt.yticks(np.arange(1.0, 3.8, step=0.2))
plt.xticks(np.arange(2, 9, step=2))
ax.plot(x, z1, 'o-', linewidth=1.5, color=(1, 0.1, 0.1), label='n = 400', markerfacecolor='white', markeredgewidth=1.5)
ax.plot(x, z2, '^--', linewidth=1.5, color=(0.1, 0.8, 0.1), label='n = 500', markerfacecolor='white', markeredgewidth=1.5)
ax.plot(x, z3, 's:', linewidth=1.5, color=(0.1, 0.1, 1.0), label='n = 600', markerfacecolor='white', markeredgewidth=1.5)
ax.plot(x, z4, 'p-.', linewidth=1.5, color=(1.0, 0.6, 0.0), label='n = 700', markerfacecolor='white', markeredgewidth=1.5)
ax.grid(axis='y', linewidth= 1)
#ax.grid()
ax.legend(loc="lower right")
```
